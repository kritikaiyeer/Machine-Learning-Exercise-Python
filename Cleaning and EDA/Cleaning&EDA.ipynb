```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('Book 4 - Sheet1.csv')
df.head()
```

```
df.info()
```

```
df.drop_duplicates(inplace=True)
df
```

```
df.drop( ['Age_Group.1'] , axis = 1, inplace=True)
```

```
df.CustomerID.loc[df.CustomerID < 0]  = np.nan
```

```
df.Bill.loc[df.Bill < 0]  = np.nan
df
```

```
df.EstimatedSalary.loc[df.EstimatedSalary < 0]  = np.nan
df
```

```
df['Rating(1-5)'].iloc[(df['Rating(1-5)'] < 0)  |  (df['Rating(1-5)'] > 5) ]  = np.nan
df
```

```
df['NoOfPax'].iloc[  (df['NoOfPax'] < 1)  |  (df['NoOfPax'] > 20) ]  = np.nan
df
```

```
df.Age_group.unique()
```

```
df.Hotel.unique()
```

```
# Replacing Ibys with Ibis
df.Hotel.replace(['Ibys'],'Ibis' , inplace=True)
df
```

```
df.FoodPreference.unique()
```

```
df.FoodPreference.replace(['Vegetarian','veg'],'Veg' , inplace=True)
df.FoodPreference.replace(['non-veg'], 'Non-Veg', inplace=True)
df
```

```
df.isna().sum()
```

```
import missingno as msno

max_salary = df["EstimatedSalary"].max()
print("\n The maximum Estimated salary = " , str(max_salary))
missingdata_df = df.columns[df.isnull().any()].tolist()
msno.matrix(df[missingdata_df])
```

```
df.NoOfPax.median()[0]
```

```
df['Rating(1-5)'].fillna( df['Rating(1-5)'].median()[0] , inplace=True )
df
```

```
df['NoOfPax'].fillna( df['NoOfPax'].median()[0] , inplace=True )
df
```

```
df.EstimatedSalary.fillna( round( df.EstimatedSalary.mean() ) , inplace=True )
df
```

```
max_salary = df["EstimatedSalary"].max()
print("\n The maximum Estimated salary = " , str(max_salary))
```
