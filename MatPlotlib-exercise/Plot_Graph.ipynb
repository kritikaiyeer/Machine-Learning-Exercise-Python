```
import numpy  as np
import pandas as pd
import matplotlib.pyplot  as plt
import matplotlib.cm      as cm
import seaborn            as sns
```

```
df = pd.read_csv('StudentsPerformance.csv', encoding='utf-8')
df.head()
```

```
df.shape
```

```
df.describe()
```

```
df.isnull().sum()
```

```
#rename coloumn
df.rename(columns = {'race/ethnicity': 'race'}, inplace = True)
```

```
# create average column
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df
```

```
#Histogram 
plt.hist(df['average score'], color='green', edgecolor= 'black')
plt.title("Average Score")
```

```
#Bar plot seaborn
bar = sns.barplot(x='race', y='reading score', data=df)
plt.xlabel('Race')
plt.ylabel('Reading Score')
plt.title('Reading Score with Race')
```

```
#Scatter Plot Seaborn
sns.scatterplot(data=df, x='math score', y='reading score')
plt.title("Reading Score with Math Score")
plt.xlabel("Math Score")
plt.ylabel('Reading Score')
```

```
sns.scatterplot(data=df, x='writing score', y='reading score', hue='gender',
               style='gender')
```

```
#Pie Plot 
labels = df.race.value_counts().index
colors = ['red', 'green', 'blue','yellow', 'orange']
explode = [0.1,0,0,0,0]
sizes = df.race.value_counts().values

# visulize

plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors)
plt.title('Race', color='black', fontsize=15)
```

```
#Line Plot
sns.lineplot(data=df, x='math score', y='reading score')
```

```
#Count Plot
plot = sns.countplot(data=df, x='gender')
plt.title("Gender Count")
```

```
#Heatmap
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt='f' ,cmap='inferno')
plt.show()
```
